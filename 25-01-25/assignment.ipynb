```python
import pandas as pd
data = pd.read_csv("GlobalLandTemperaturesByCity.csv")
data
```

```python
data['dt'] = pd.to_datetime(data['dt'])
```

```python
data.info()
```

```python
data['AverageTemperature'] = data['AverageTemperature'].fillna(data['AverageTemperature'].mean())
```

```python
data
```

```python
data['AverageTemperatureUncertainty'] = data['AverageTemperatureUncertainty'].fillna(data['AverageTemperatureUncertainty'].mean())
```

```python
data
```

```python
data['Year'] = data['dt'].dt.year
```

```python
data
```

```python
city_data = data[data['City'] == 'Bordeaux']
city_data
```

```python
import matplotlib.pyplot as plt
annual_stats = city_data.groupby('Year')['AverageTemperature'].agg(['mean', 'min', 'max']).reset_index()
annual_stats
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(annual_stats['Year'], annual_stats['mean'], color='blue', label='Average Temp', alpha=0.8)
plt.scatter(annual_stats['Year'], annual_stats['min'], color='green', label='Min Temp', alpha=0.6)
plt.scatter(annual_stats['Year'], annual_stats['max'], color='red', label='Max Temp', alpha=0.6)
```
